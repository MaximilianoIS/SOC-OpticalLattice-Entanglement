```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("contrast_oat.csv")

plt.figure(figsize=(7,5),dpi=300)
plt.plot(df['t'], df['Cz'], label=r'$C^{(Z)}$', color='orange')
plt.xlabel("t")
plt.ylabel(r"$C^{(Z)}$")
plt.title("OAT Contrast $C^{(Z)}$ vs t")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("contrast_oat.csv")

df_filtered = df[df['t'] <= 200]

plt.figure(figsize=(7, 5), dpi=300)
plt.plot(df_filtered['t'], df_filtered['Cz'], label=r'$C^{(Z)}$', color='orange')
plt.xlabel("t")
plt.ylabel(r"$C^{(Z)}$")
plt.title("OAT Contrast $C^{(Z)}$ vs t (t ≤ 200)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
